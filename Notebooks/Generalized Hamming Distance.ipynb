# Generalized Hamming Distance
This notebook explores the fundamental building block of the Generalized Hamming Network which is based on the **Generalized Hamming Distance**.

We will explore and comfirm some of the definitions more concretely through code implementations for each of the definition.

## Definition 1
Let $a, b, c \in U \subseteq \mathbb{R}$, and a *Generalized Hamming Distance (GHD)*, denoted by $\oplus$, be a binary operator $h: U \times U \longrightarrow U$.

$h(a, b) := a \oplus b = a + b = 2 \cdot a \cdot b$

```python
def ghd(a, b):
    return a + b - 2 * np.multiply(a, b)
```

### Property 1:
For $U = \{0, 1\}$ GHD de-generalizes to binary hamming distance with:

$0 \oplus 0 = 0$

$0 \oplus 1 = 1$

$1 \oplus 0 = 1$

$1 \oplus 1 = 0$

```python
print('ghd(0, 0): ' + str(ghd(0, 0)))
print('ghd(1, 0): ' + str(ghd(1, 0)))
print('ghd(0, 1): ' + str(ghd(0, 1)))
print('ghd(1, 1): ' + str(ghd(1, 1)))
```

## Plotting the GHD Function

### Figure 1: $h(a, b)$

```python
fig_1 = plt.figure()
ax_1 = fig_1.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 4.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([ghd(x, y) for x, y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax_1.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.show()
```

### Figure 2: Fuzziness Function - $F(h(a, b)) = h(a, b) \oplus h(a, b)$

```python
fig_1 = plt.figure()
ax_2 = fig_1.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 4.0, 0.01)
X, Y = np.meshgrid(x, y)
zs = np.array([ghd(ghd(x, y), ghd(x, y)) for x, y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax_2.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.show()
```

### Figure 3: $\mu(h(a,b)) = \dfrac{1}{1+exp(0.5-h(a,b))}$

```python
mu = lambda x: 1 / (1 + np.exp(0.5-x))

fig_1 = plt.figure()
ax_2 = fig_1.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 4.0, 0.01)
X, Y = np.meshgrid(x, y)
zs = np.array([mu(ghd(x, y)) for x, y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax_2.plot_surface(X, Y, Z, cmap=cm.coolwarm)
plt.show()
```
